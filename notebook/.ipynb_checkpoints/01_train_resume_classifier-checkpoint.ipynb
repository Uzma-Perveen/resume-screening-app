```python
import re, json, string
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import joblib

import nltk
nltk.download('stopwords')
nltk.download('wordnet')
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

STOPWORDS = set(stopwords.words('english'))
LEMMATIZER = WordNetLemmatizer()
```

```python
df = pd.read_csv("../data/resume_dataset.csv")
df = df.rename(columns={c:c.strip() for c in df.columns})
df = df.dropna(subset=["Category", "Resume"]).reset_index(drop=True)

print(df.head(2))
print(df["Category"].value_counts())
```

```python
URL_RE = re.compile(r'https?://\S+|www\.\S+')
EMAIL_RE = re.compile(r'\S+@\S+')
NUM_RE = re.compile(r'\b\d+\b')

def clean_text(text: str) -> str:
    text = text.lower()
    text = URL_RE.sub(' ', text)
    text = EMAIL_RE.sub(' ', text)
    text = NUM_RE.sub(' ', text)
    text = text.translate(str.maketrans('', '', string.punctuation))
    tokens = [LEMMATIZER.lemmatize(tok) for tok in text.split() if tok not in STOPWORDS and len(tok) > 2]
    return " ".join(tokens)

df["clean_text"] = df["Resume"].astype(str).apply(clean_text)
df = df[df["clean_text"].str.len() > 0]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    df["clean_text"], df["Category"], test_size=0.2, random_state=42, stratify=df["Category"]
)
```

```python
pipe_lr = Pipeline([
    ("tfidf", TfidfVectorizer(ngram_range=(1,2), min_df=3, max_df=0.9, sublinear_tf=True)),
    ("clf", LogisticRegression(max_iter=2000, n_jobs=None, class_weight=None))
])

pipe_lr.fit(X_train, y_train)
pred_lr = pipe_lr.predict(X_test)
proba_lr = pipe_lr.predict_proba(X_test)
print("LogReg Accuracy:", accuracy_score(y_test, pred_lr))
print(classification_report(y_test, pred_lr))
```

```python
pipe_nb = Pipeline([
    ("tfidf", TfidfVectorizer(ngram_range=(1,2), min_df=3, max_df=0.9, sublinear_tf=True)),
    ("clf", MultinomialNB())
])

pipe_nb.fit(X_train, y_train)
pred_nb = pipe_nb.predict(X_test)
proba_nb = pipe_nb.predict_proba(X_test)
print("Naive Bayes Accuracy:", accuracy_score(y_test, pred_nb))
print(classification_report(y_test, pred_nb))
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib.pyplot as plt

best_pipe = pipe_lr  # or pipe_nb
pred_best = best_pipe.predict(X_test)

cm = confusion_matrix(y_test, pred_best, labels=best_pipe.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=best_pipe.classes_)
fig, ax = plt.subplots(figsize=(8,6))
disp.plot(ax=ax, xticks_rotation=45, cmap="Blues")
plt.title("Confusion Matrix - Resume Role Classification")
plt.show()
```

```python
# Fit a fresh TF-IDF on full cleaned text for feature names
tfidf = TfidfVectorizer(ngram_range=(1,2), min_df=5, max_df=0.95, sublinear_tf=True)
X_all = tfidf.fit_transform(df["clean_text"])
labels = df["Category"].values

# Get top terms per class via class-conditional mean tf-idf
top_skills = {}
for role in sorted(df["Category"].unique()):
    idx = np.where(labels == role)[0]
    mean_vec = np.asarray(X_all[idx].mean(axis=0)).ravel()
    top_idx = mean_vec.argsort()[::-1][:40]
    terms = [tfidf.get_feature_names_out()[i] for i in top_idx]
    # keep only single tokens that look like hard skills (simple heuristic)
    top_skills[role] = [t for t in terms if " " not in t][:25]

with open("../models/top_skills_per_role.json", "w") as f:
    json.dump(top_skills, f, indent=2)
```
